멜론 차트 TOP100수집

```python
import requests as re
from bs4 import BeautifulSoup as bs
import pandas as pd

# row 생략 없이 출력 (전체 출력)
pd.set_option('display.max_rows', None)
# col 생략 없이 출력 (전체 출력)
pd.set_option('display.max_columns', None)

# 사람인지 속이기 h
h={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
# 주소 불러오기
res = re.get('https://www.melon.com/chart/index.htm',headers=h)
soup = bs(res.content,'html.parser')
#인기 차트
data2 = soup.select('div.ellipsis.rank01 > span > a')
#가수명
data3 = soup.select('div.ellipsis.rank02 > .checkEllipsis')
#앨범명
data4 = soup.select('.ellipsis.rank03 >  a')
#a = 인기차트.text 만 리스트에 담기
a = [i.text for i in data2]
#b = 가수명.text 만 리스트에 담기
b = [i.text for i in data3]
#c = 앨범명.text 만 리스트에 담기
c = [i.text for i in data4]
#d = 1부터 100까지 숫자 리스트에 담기
d = [i for i in range(1,101)]
#판다스 형식으로 만들어 주기
da = {
    '순위':d,
    '노래제목':a,
    '가수명':b,
    '앨범제목':c
}
#DataFrame에 넣기
df = pd.DataFrame(da)
#da 의 순위를 인덱스 로 정함
df.set_index('순위',inplace=True)
#csv 저장 
df.to_csv('music2.csv',index = False, encoding='utf-8')
df
```

#### YES24 불러오기

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

res = req.get('http://www.yes24.com/24/Category/BestSeller')
soup = bs(res.content,'html.parser')
data = soup.select("ol > li")
print(data[0])
for i in data:
    title = i.select('p>a')[2].text
    price = i.select('.price > strong')[0].text
    print(title," : ",price)
# data = soup.select('p:nth-child(3) > a')
# for i in data:
#     print(i.text)
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

res = req.get('https://finance.naver.com/sise/lastsearch2.naver')
soup = bs(res.content,'html.parser')
data = soup.select('td:nth-child(2) > a')
data2 = soup.select('td:nth-child(8)')
for i,j in zip(data,data2):
    print(i.text,' : ',j.text)
```
