```python
from selenium import webdriver  as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyperclip
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
#1번쨰
url = 'https://korean.visitkorea.or.kr'
driver = wb.Chrome()
driver.get(url)
time.sleep(2)
```

```python
#2번쨰
driver.find_element(By.XPATH,'//*[@id="placeHolder"]/a').click()
serch = driver.find_element(By.XPATH,'//*[@id="inp_search"]')
serch.send_keys('제주도')
serch.send_keys(Keys.ENTER)
```

```python
html = driver.page_source
soup = bs(html,'html.parser')
```

```python
content_list = soup.find('ul','list_thumType type1')
coun = 1
for i in content_list:
    print(coun,i.text.strip())
    coun+=1
```

```python
f = open('제주도.txt','w',encoding='utf-8')
content_list = soup.find('ul','list_thumType type1')
coun = 1
for i in content_list:
    f.write(str(coun))
    f.write(i.text.strip())
    f.write('\n\n')
    coun+=1
f.close()
```

```python
#3번쨰
count = 100
one_page_data = 10
need_page_count=10
no=1     #게시글 번호 카운트
no_list=[]  # 게시글 번호 저장용리스트
contents_list=[]  #게시글 제목 저장용 리스트

for page in range(1,need_page_count+1): #1부터 10페이지까지 
    print("="*100)
    print(f"{page}페이지 내용 크롤링 시작합니다")
    print("-"*75)
    time.sleep(2)
    
    html=driver.page_source
    soup=bs(html,'html.parser')
    
    content_list=soup.find('ul','list_thumType type1').find_all('div','area_txt')
    
    co=0
    for i in content_list:
        print("🔸🔸🔸",content_list[co].text)
        co+=1
        #게시글 번호 
        print("🔵🔵번호:",no)
        no_list.append(no)
        no+=1
        #게시글 제목 
        contents=i.find('div','tit').text.strip()
        print("제목:",contents)
        contents_list.append(contents)
        
    next_page=page+1
    print("next page:",next_page)
#     driver.find_element_by_link_text(f'{page+1}').click()
    try :      
        if (page%5) == 0 :
            #driver.find_element(By.LINK_TEXT,"다음").click()
            driver.find_element_by_link_text("%s" %"다음").click()  
            time.sleep(3)
        else :  
            #html 코드를 보면 페이지를 의미하는1,2,3,... 부분의 이름이 1,2,3...으로 사용되고 있으므로 page 값으로 사용한다.  
            driver.find_element_by_link_text("%s" %next_page).click()  
             #driver.find_element(By.LINK_TEXT,next_page).click()
            
    except :        
        print("다음 페이지가 없어서 자동 종료합니다.")
        break    
```
