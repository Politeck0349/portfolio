### 네이버 주식 현재가 정보 수집하기
- 네이버 -> 증권 -> 삼성전자입력 -> 시세
- https://finance.naver.com/item/sise.naver?code=005930
    - 종목 코드 : 005930

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
url = 'https://finance.naver.com/item/sise.naver?code=130660'
res = req.get(url)
soup = bs(res.content, 'html.parser')
price = soup.select_one('#_nowVal').text
price.replace(',','')
```

```python
#삼성전자 카카오 코오롱인더, sk하이닉스, 현대차
codes=['005930','035720','120110','000660','005380']
#반복문사용하여 5종목 시세가져오기 
for code in codes:
    url=f"https://finance.naver.com/item/sise.naver?code={code}"
    res=req.get(url)
    soup=bs(res.content,'lxml')
    price=soup.select_one("#_nowVal").text
    price=price.replace(",","")
    print(price)
```

```python
#엑셀열기
import openpyxl
#1.엑셀파일 만들기 
wb=openpyxl.Workbook()
#2.엑셀 워크시트만들기
ws=wb.create_sheet("네이버주식")
#3.데이터 추가하기 
ws['A1']='종목'
ws['B1']='현재가'
ws['A2']='삼성전자'
ws['A3']='카카오'
ws['A4']='코오롱인더'
ws['A5']='sk하이닉스'
ws['A6']='현대차'
#삼성전자 카카오 코오롱인더, sk하이닉스, 현대차
codes=['005930','035720','120110','000660','005380']
#반복문사용하여 5종목 시세가져오기 
row=2
for code in codes:
    url=f"https://finance.naver.com/item/sise.naver?code={code}"
    res=req.get(url)
    soup=bs(res.content,'lxml')
    price=soup.select_one("#_nowVal").text
    price=price.replace(",","")
    print(price)
    ws[f'B{row}'] = int(price)
    row+=1
#엑셀 저장하기 
wb.save('stock.xlsx')
```
